```python
import matplotlib.pyplot as plt
import networkx as nx
from leave_nodes import balanced_tree_with_fixed_leaves, num_leaves

import warnings

warnings.simplefilter(action="ignore", category=DeprecationWarning)
```

```python
num_workers = 512
heights = [1, 2, 4, 16]
fig, axes = plt.subplots(nrows=1, ncols=len(heights))

for ax, h in zip(axes, heights):
    print(f"Making tree with height {h}...")
    tree = balanced_tree_with_fixed_leaves(num_workers, h)
    pos = nx.nx_pydot.pydot_layout(tree, prog="dot")
    nx.draw(tree, node_size=10, ax=ax, pos=pos)
    count = num_leaves(tree)
    ax.set_title(f"{count} leaves")

plt.tight_layout()
plt.show()
```
